```python
# TASK 3: DATA VISUALIZATION (COMPLETE CODE)
# ==========================================

# 1. Import required libraries for data manipulation and visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the Seaborn theme for clean and visually appealing plots
sns.set_theme(style="whitegrid")

# 2. Create a dummy dataset representing Sales & Profit Data
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Sales_USD': [12000, 15000, 18000, 14000, 22000, 26000, 21000, 24000, 29000, 31000, 42000, 48000],
    'Profit_USD': [2000, 3100, 4200, 2500, 5000, 6200, 4900, 5600, 7100, 8000, 11500, 13000],
    'Category': ['Electronics', 'Clothing', 'Electronics', 'Home Decor', 'Clothing', 'Electronics',
                 'Home Decor', 'Clothing', 'Electronics', 'Home Decor', 'Electronics', 'Clothing']
}

# Load data into a pandas DataFrame
df = pd.DataFrame(data)
print("--- Data Loading Successful! ---")
print(df.head()) # Display the first 5 rows of the dataset
print("\n")

# --------------------------------------------------
# VISUAL 1: Line Plot (Monthly Sales and Profit Trend)
# --------------------------------------------------
plt.figure(figsize=(10, 5))

# Plotting Sales and Profit trends across the months
sns.lineplot(x='Month', y='Sales_USD', data=df, marker='o', color='b', linewidth=2.5, label='Sales')
sns.lineplot(x='Month', y='Profit_USD', data=df, marker='s', color='g', linewidth=2.5, label='Profit')

# Enhancing visual clarity with titles and labels
plt.title('Monthly Sales & Profit Trend (Year 2026)', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Months', fontsize=12)
plt.ylabel('Amount in USD ($)', fontsize=12)
plt.legend(loc='upper left')
plt.tight_layout()
plt.show() # Render the first visualization

# --------------------------------------------------
# VISUAL 2: Bar Plot (Revenue Generated by Category)
# --------------------------------------------------
plt.figure(figsize=(8, 5))

# Grouping data by Category to aggregate total sales
category_sales = df.groupby('Category')['Sales_USD'].sum().reset_index()

# Creating a bar chart using Seaborn's magma palette
sns.barplot(x='Category', y='Sales_USD', data=category_sales, palette='magma')

# Customizing the bar chart
plt.title('Total Revenue Generated by Product Category', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Product Category', fontsize=12)
plt.ylabel('Total Sales ($)', fontsize=12)
plt.tight_layout()
plt.show() # Render the second visualization

# --------------------------------------------------
# VISUAL 3: Scatter Plot (Sales vs Profit Relationship)
# --------------------------------------------------
plt.figure(figsize=(8, 5))

# Plotting the correlation between Sales and Profit segmented by Category
sns.scatterplot(x='Sales_USD', y='Profit_USD', hue='Category', style='Category', data=df, s=100)

# Finalizing the plot details
plt.title('Relationship Between Sales and Profit', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Sales (USD)', fontsize=12)
plt.ylabel('Profit (USD)', fontsize=12)
plt.tight_layout()
plt.show() # Render the third visualization
```
